```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("archive2/train.txt",header=None,sep=';').values
test_data = pd.read_csv("archive2/test.txt",header=None,sep=';').values
```

```python
train_data[:10]
```

```python
import matplotlib.pyplot as plt
```

```python
class_name, class_count = np.unique(train_data[:,1], return_counts=True)
plt.figure(figsize=(8, 6))
plt.bar(class_name, class_count, color=['red', 'purple', 'green', 'pink', 'blue', 'orange'])
plt.title('Emotion Distribution')
plt.xlabel('Emotion')
plt.ylabel('Count')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
class_name, class_count = np.unique(test_data[:,1], return_counts=True)
plt.figure(figsize=(8, 6))
plt.bar(class_name, class_count, color=['red', 'purple', 'green', 'pink', 'blue', 'orange'])
plt.title('Emotion Distribution')
plt.xlabel('Emotion')
plt.ylabel('Count')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
